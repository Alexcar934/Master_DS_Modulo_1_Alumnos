## 2. Preparaci칩n de Datos <a name="preparacion"></a>

Para ilustrar estas operaciones, creemos un DataFrame de ejemplo. (En un caso real, podr칤amos usar `pd.read_csv()` u otras funciones para cargar datos de un archivo.)

En este ejemplo, generaremos un DataFrame de ventas con las siguientes columnas:
- **Producto**
- **Precio Unitario**
- **Unidades Vendidas**
- **Descuento (%)**

---


```python
    
datos_ventas = pd.read_csv('supermarket_sales - Sheet1.csv')

df = pd.DataFrame(datos_ventas)
df
```

## 2.1 Renombrar solo la columna 'cogs'

Vamos a renombrar 칰nicamente la columna `cogs` para que sea m치s clara: `Cost of Goods Sold`.


```python
# Renombrar solo la columna 'cogs'
df.rename(columns={"cogs": "Cost of Goods Sold"}, inplace=True)

# Verificamos que se haya renombrado correctamente
print("Nombres de columnas despu칠s del renombrado:")
print(df.columns)
```

## 2.2 Crear nuevas columnas: Subtotal e Impuesto Calculado

Aunque el dataset ya trae columnas como 'Tax 5%' y 'Total', vamos a calcular el 'Subtotal' restando el impuesto, 
y comprobar si el impuesto realmente es el 5%.


```python
# Subtotal = Total - Tax 5%
df["Subtotal"] = df["Total"] - df["Tax 5%"]

# Impuesto calculado (en porcentaje)
df["Impuesto Calculado (%)"] = (df["Tax 5%"] / df["Subtotal"]) * 100

# Mostramos una vista general
df[["Invoice ID", "Unit price", "Quantity", "Subtotal", "Tax 5%", "Total", "Impuesto Calculado (%)"]]
```

## 2.3 Trabajando con fechas y horas

Convertimos las columnas de fecha y hora a tipos datetime para extraer informaci칩n 칰til como a침o, mes, d칤a, hora, etc.


```python
df.head()
```

```python
# Convertimos las columnas Fecha y Hora
df["Date"] = pd.to_datetime(df["Date"])
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M")

# Extraemos partes 칰tiles de fecha y hora
df["A침o"] = df["Date"].dt.year
df["Mes"] = df["Date"].dt.month
df["D칤a de la Semana"] = df["Date"].dt.day_name()
df["Hora del D칤a"] = df["Time"].dt.hour

# Vista r치pida
df[["Invoice ID", "Date", "Time", "A침o", "Mes", "D칤a de la Semana", "Hora del D칤a"]]
```

## 2.4 Clasificaci칩n con `apply`: Calificaci칩n (Rating)

Creamos una nueva columna que clasifica las calificaciones en:
- "Alta" si es mayor o igual a 9
- "Media" si est치 entre 7 y 9
- "Baja" si es menor a 7


```python
def clasificar_rating(r):
    if r >= 9:
        return "Alta"
    elif r >= 7:
        return "Media"
    else:
        return "Baja"

df["Nivel de Calificaci칩n"] = df["Rating"].apply(clasificar_rating)

df[["Invoice ID", "Rating", "Nivel de Calificaci칩n"]]
```

```python
# Crear una columna con lambda

df['is_Yangon'] = df['City'].apply(lambda x: True if x == 'Yangon' else False)
```

## 游빞 Limpieza de datos: duplicados y valores nulos

En esta secci칩n aprenderemos a usar algunos de los m칠todos m치s comunes de pandas para **limpiar un DataFrame**:

---

### 游대 Duplicados

- `duplicated()`  
Devuelve una serie booleana que indica si una fila est치 duplicada respecto a una anterior.

- `drop_duplicates()`  
Elimina filas duplicadas. Se puede usar sobre todas las columnas o sobre un subconjunto espec칤fico.



```python
df.columns
```

```python
df_short = df[['Invoice ID', 'Gender','Total','Rating', 'City']] 

df_with_dup = pd.concat([df_short.iloc[:500,:],df_short],axis=0)

filas_nulas = pd.DataFrame([
    {"Invoice ID": None, "Gender": None, "Total": np.nan, "Rating": np.nan}, 
    {"Invoice ID": "000-00-0000", "Gender": "Female", "Total": 300.0, "Rating": np.nan},
    {"Invoice ID": "111-11-1111", "Gender": None, "Total": None, "Rating": 9.0}
])

# pandas rellenar치 con NaN las columnas que falten al unir con df
df_with_dup_and_na = pd.concat([df_with_dup, filas_nulas], axis=0, ignore_index=True)
```

```python
df_with_dup_and_na
```

```python
# 游대 Verificar si hay filas duplicadas
# True indica que esa fila ya ha aparecido antes con los mismos valores
duplicados = df_with_dup_and_na.duplicated()
print("쮿ay filas duplicadas?")
print(duplicados.value_counts())
# print(df['City'].duplicated())
```

```python
df_with_dup_and_na[df_with_dup_and_na.duplicated()==True]
```

```python
# 游대 Eliminar duplicados
# Borra las filas duplicadas y devuelve un nuevo DataFrame (no modifica el original a menos que uses inplace=True)
df_sin_duplicados = df_with_dup_and_na.drop_duplicates()
print(f"N칰mero de filas despu칠s de eliminar duplicados: {len(df_sin_duplicados)}")
```

---

### 仇 Valores nulos

- `isna()` o `isnull()`  
Devuelve `True` donde hay valores nulos en el DataFrame.

- `dropna()`  
Elimina filas o columnas que contienen valores nulos.

- `fillna(valor)`  
Permite rellenar los valores nulos con un valor espec칤fico (por ejemplo, 0 o la media de una columna).

---

Estos m칠todos son fundamentales para preparar los datos antes de analizarlos o visualizarlos.


```python
# 仇 Buscar valores nulos en todo el DataFrame
nulos = df_with_dup_and_na.isna().sum()
print("Valores nulos por columna:")
print(nulos)
```

```python
# 仇 Eliminar filas con cualquier valor nulo
# 丘멆잺 Cuidado: esta operaci칩n puede eliminar muchas filas si no se usa con criterio
df_sin_nulos = df_with_dup_and_na.dropna()
print(f"N칰mero de filas tras eliminar nulos: {len(df_sin_nulos)}")
```

```python
# 仇 Rellenar valores nulos con un valor espec칤fico
# Ejemplo: rellenamos nulos en "Rating" con la media
media_rating = df_with_dup_and_na["Rating"].mean()
df_with_dup_and_na["Rating"] =df_with_dup_and_na["Rating"].fillna(media_rating)



print("Comprobamos si quedan nulos en Rating:")
print(df_with_dup_and_na["Rating"].isna().sum())
```
